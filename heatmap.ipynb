```python
import pandas as pd
globalWarming_df=pd.read_csv("global.csv")
globalWarming_df.head()
```

```python
globalWarming_df=globalWarming_df.drop(columns=['Country Code',
                                               'Indicator Name',
                                               'Indicator Code'],
                                      axis=1).set_index('Country Name')
globalWarming_df
```

```python
plt.figure(figsize=(16,20))
sns.heatmap(globalWarming_df)
```

```python
plt.figure(figsize=(26,20))
sns.heatmap(globalWarming_df,vmin=0,vmax=21)
```

```python
plt.figure(figsize=(16,20))
sns.heatmap(globalWarming_df,cmap="coolwarm")
```

```python
plt.figure(figsize=(16,20))
sns.heatmap(globalWarming_df,cmap="coolwarm",center=0.5)
```

```python
plt.figure(figsize=(16,20))
sns.heatmap(globalWarming_df,annot=True)
```
